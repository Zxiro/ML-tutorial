# 動手做

修改範例程式，試著使用不同的機器學習演算法，建構比範例分類器更好的模型。

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier # import required classifiers
import sys
sys.path.append('.prepared')
import classification as prepared

X_train, X_test, y_train, y_test = train_test_split(
    prepared.X,
    prepared.y,
    random_state = 0, # try a different number
    test_size = 0.1 # try a different ratio
)

demo_clf = prepared.classifier(X_train, y_train)
prepared.evaluate(demo_clf, X_train, X_test, y_train, y_test) # beat the results

clf = DecisionTreeClassifier(max_depth=2) # try a different classifier
clf = clf.fit(X_train, y_train)
prepared.evaluate(clf, X_train, X_test, y_train, y_test)

if 'DecisionTreeClassifier' == type(clf).__name__:
    prepared.plot(clf)
```
